```python
import pandas as pd
```

```python
air = pd.read_csv("death-rates-total-air-pollution.csv")
water = pd.read_csv("death-rates-unsafe-water.csv")
waste = pd.read_csv("plastic-waste-per-capita.csv")
```

```python
air.head()
```

```python
water.head()
```

```python
waste.head()
```

```python
air = air[air.Year == 2017].reset_index(drop=True)
water = water[water.Year == 2017].reset_index(drop=True)
waste = waste[waste.Year == 2010].reset_index(drop=True)
```

```python
merged = air.merge(water, left_on="Entity", right_on="Entity").merge(waste, left_on="Entity", right_on="Entity")
```

```python
merged
```

```python
merged = merged[['Entity', 'Year', 'Year_x', 'Code', 'Deaths - Air pollution - Sex: Both - Age: Age-standardized (Rate)',
                'Deaths - Unsafe water source - Sex: Both - Age: Age-standardized (Rate)',
                'Per capita plastic waste (kg/person/day)']]
```

```python
merged
```

```python
country = pd.read_csv("countries_lat_lon.csv")
```

```python
country.head()
```

```python
final = merged.merge(country, left_on="Entity", right_on="name")
```

```python
final.to_csv("country_pollution.csv")
```
